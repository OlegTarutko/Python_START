```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
```

```python
f = x**3 - 50*x
g = -x**4 + 88*x**2 - 241
```

```python
#Координаты точек
k = solve(f - g)
[[xelem.evalf().round(2),f.subs(x,xelem).evalf().round(2)] for xelem in k]
```

```python
#Функции в одной системе координат.
twofunctions = plot(f,g)
```

```python
#График пересечения.
intersection = plot(f - g)
```
